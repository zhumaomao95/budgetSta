```python
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据
np.random.seed(0)
x = np.arange(10)
y = np.random.randint(1, 20, size=10)

# 配色方案
colors = plt.cm.viridis(np.linspace(0, 1, len(x)))

# 绘制柱状图
plt.bar(x, y, color=colors)

# 添加标题和标签
plt.title('Random Bar Chart with Nice Colors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 显示图形
plt.show()
```
